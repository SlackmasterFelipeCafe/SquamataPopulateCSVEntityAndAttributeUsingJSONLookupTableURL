```python
import json
import requests
import pandas as pd
import sciencebasepy as pysb
from io import StringIO
```

```python
#login to ScienceBase
sb = pysb.SbSession()
username = input("Username:  ")
sb.loginc(str(username))
#Check to see if login is successful
sb.is_logged_in()
#Get the ScienceBase Login session info
sb.get_session_info()
```

```python
# Now lets get the item - be sure to use the correct catalog number!!!
catNum = '621ce846d34ec739b2de39d9'
item_json = sb.get_item(catNum)
#print ("Public Item: \n\t" + str(item_json))
print(json.dumps(item_json, indent=2, sort_keys=True))
```

```python
for item in item_json['files']:
 print(item['name'])
 print(item['url'])
    
strURL_JSON = item['url']
```

```python
url = requests.get(strURL_JSON)
text = url.text
jsonData = json.loads(text)
pretty_object = json.dumps(jsonData, indent=4)
print(pretty_object)
```

```python
print(jsonData["Attrlabl"])
```

```python
# Create arrays of JSON data to be searched and inserted into metadata.
##print(jsonData['fields']['ALT']['description'])
arrAttrlabl = []
arrAttrdef = []
arrAttrunit = []


for key, value in jsonData['Attrlabl'].items():
    print(key, value)
    try:
     arrAttrlabl.append(value)
    except:
          print("\t\t !!! NO Value Found In Lookup Table !!!!")
          arrAttrlabl.append('!!! NO Value Found In Lookup Table !!!!')
          continue

for key, value in jsonData['Attrdef'].items():
    print(key, value)
    try:
     arrAttrdef.append(value)
    except:
          print("\t\t !!! NO Value Found In Lookup Table !!!!")
          arrAttrdef.append('!!! NO Value Found In Lookup Table !!!!')
          continue

for key, value in jsonData['Attrunit'].items():
    print(key, value)
    try:
     arrAttrunit.append(value)
    except:
          print("\t\t !!! NO Value Found In Lookup Table !!!!")
          arrAttrunit.append('!!! NO Value Found In Lookup Table !!!!')
          continue

#for i in range(len(arrAttrlabl)):
 #print (str(i) + ',' + arrAttrlabl[i] + ',' + arrAttrdef[i] + ',' + arrAttrunit[i])
```

```python
#Lets get the file we want to work on:
for item in item_json['files']:
 print(item['name'])
 print(item['url'])
    
strURL_JSON = item['url']
```

```python
#Set Data Paths to a SB URL

emriFilePath = 'https://www.sciencebase.gov/catalog/file/get/63a4c31ed34efb1ef7636307?f=__disk__19%2Fb5%2F95%2F19b595b4c13cbfc7e7fdb71231be6180f367724c'
```

```python
# read file into pandas; note that can take over 1 minute depending on the size of the CSV file and computing power.
print ('!!! PLEASE BE PATIENT AND TAKE SEVERAL BREATHS !!!')
print ('Reading file into Pandas, this may take several minutes depending on the file size...')

getURL = requests.get(emriFilePath
csvStringIO = StringIO(getURL.text)
emriDataFrame = pd.read_csv(csvStringIO, sep=',', delimiter=None, header='infer', names=None, index_col=None, usecols=None, squeeze=False, prefix=None, mangle_dupe_cols=True, dtype=None, engine=None, converters=None, true_values=None, false_values=None, skipinitialspace=False, skiprows=None, nrows=None, na_values=None, keep_default_na=True, na_filter=True, verbose=False, skip_blank_lines=True, parse_dates=False, infer_datetime_format=False, keep_date_col=False, date_parser=None, dayfirst=False, iterator=False, chunksize=None, compression='infer', thousands=None, decimal=b'.', lineterminator=None, quotechar='"', quoting=0, doublequote=True, escapechar=None, comment=None, encoding=None, dialect=None, error_bad_lines=True, warn_bad_lines=True, delim_whitespace=False, low_memory=True, memory_map=False, float_precision=None)
emriDataFrame
```

```python
# now get an array of the header names
emriHeaderList = list(emriDataFrame.columns.values)
emriHeaderString = ''
for i in range(len(emriHeaderList)):
 emriHeaderString = emriHeaderString + emriHeaderList[i] + ','

print (emriHeaderString)
```

```python
print(len (arrAttrlabl))
print(len (arrAttrdef))
print(len (arrAttrunit))
```

```python
arrDef = []
arrUnits = []
arrHdrFound = []
for i in range(len(emriHeaderList)):
    for j in range(len(arrAttrlabl)):
        if emriHeaderList[i] == arrAttrlabl[j]:
            print(emriHeaderList[i] + '==' + arrAttrlabl[j])
            if arrAttrdef[j] != None:
             arrDef.append(arrAttrdef[j])
            if arrAttrdef[j] == None:
             arrDef.append("!!!Definition Not Defined In Lookup Table!!!")
            if arrAttrunit[j] != None:
             arrUnits.append(arrAttrunit[j])
            if arrAttrunit[j] == None:
             arrUnits.append("!!!Definition Not Defined In Lookup Table!!!")
            arrHdrFound.append(arrAttrlabl[j])
           
for k in range(len(arrDef)):
 try: 
  print (str(k) + ',' + emriHeaderList[k] + ',' + arrDef[k] + ',' + arrUnits[k])
 except:
  print(str(k) + "\t\t !!! can only concatenate str (not NoneType) to str !!!!")
```

```python
#lets see if we are missing any?

for k in range(len(emriHeaderList)):
 #print (emriHeaderList[k])
 c = 0
 for j in range(len(arrHdrFound)):
  if emriHeaderList[k] == arrHdrFound[j]:
   print (emriHeaderList[k]  + ' header found, definition:' + arrDef[j])
   c = 1
 if c==0:  
   print (emriHeaderList[k] + " !!!No Header Match Found!!!!")
```

```python
#Now create EandA for headers we have definitions for
#If we don't have definitions - these need to be added to the JSON list
csvHeaderListLength = len(arrHdrFound)
csvMax = []
csvMin = []
Attrdefs = 'EON GEOSCIENCES INC.'
entAttText = ''
#print ('Length of Entity Description Array: ' + str(len(csvEntityDescription)))
#print ('Length of Header List: ' + str(len(magHeaderList)))

for i in range(csvHeaderListLength): #csvHeaderListLength
 print (arrHdrFound[i] + ': ' + str(i))
 strCol = arrHdrFound[i]
 csvMax.append(emriDataFrame[strCol].max())
 csvMin.append(emriDataFrame[strCol].min())
 print ('Max. Value ' + arrHdrFound[i] + ': ' + str(csvMax[i]))
 print ('Min. Value ' + arrHdrFound[i] + ': ' + str(csvMin[i]))
 print ('Entt Desc: ' + arrDef[i])
 print ('Att Unit: ' + arrUnits[i] + '\n')
 entAtti = '\n\t\t\t\t<attr>\n\t\t\t\t\t<attrlabl>' + strCol + '</attrlabl>\n' \
        + '\t\t\t\t\t<attrdef>' + arrDef[i] + '</attrdef>' \
        + '\n\t\t\t\t\t<attrdefs>' + Attrdefs + '</attrdefs>' \
        + '\n\t\t\t\t\t<attrdomv>\n\t\t\t\t\t\t<rdom>\n\t\t\t\t\t\t\t<rdommin>'+ str(csvMin[i]) + '</rdommin>\n\t\t\t\t\t\t\t<rdommax>' \
        + str(csvMax[i]) + '</rdommax>\n\t\t\t\t\t\t\t' + '<attrunit>' + arrUnits[i] + '</attrunit>\n\t\t\t\t\t\t</rdom>\n\t\t\t\t\t\t' \
        + '</attrdomv>\n' \
        + '\t\t\t\t</attr>'
 entAttText = entAttText + entAtti
    
entAttText = entAttText  + '\n\t\t</detailed>\n' 

csvAttributes = entAttText

print ('\nEntity and Attribute Text:\n' + csvAttributes)
```
